Q2 : Write a program which will find all such numbers which are divisible by 7 but are not a multiple
of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a
comma-separated sequence on a single line.

```python
l = []
for i in range(2000,3201):
    if ((i % 7) == 0) and ((i % 5) != 0):
        l.append(i)

print(*l, sep = ',', end = '.') # * unwraps list element into diff arguments


```

Q1 : Write a program which accepts a sequence of comma-separated numbers from console and
generate a list.

```python
var = input('enter nos seperated by commas')
list_str = var.split(',') 
list_int = []  
len(list_str)
for i in range(0,len(list_str)):
    list_int.append(int(list_str[i]))   #type casting str(input) -> int 
    
```

```python
print(type(list_str[0]))
list_int  # list of elements of type string
```

```python
print(type(list_int[0]))
list_int      # list of elements of type int
```

Q2 : Create the below pattern using nested for loop in Python.


```python
for i in range(1,7):
    if(i<6):
        print('*'*i)
    elif (i == 6):
        for j in range(i-2,0,-1):
            print('*'*j)
```

Q3 :  Write a Python program to reverse a word after accepting the input from the user.

```python
word = input('enter the string')
rev_word = ''
for i in range(0,len(word)):
    rev_word =   rev_word + word[len(word)-i-1]
print(rev_word)
```
